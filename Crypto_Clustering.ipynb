```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
# This step centers each feature (mean = 0), and scales each feature (standard deviation = 1). 
# This step sets the data up for K-Means and PCA since they are distance-based algorithms. 
# By applying the .fit_transform directly on the DataFrame, it is possible to change the DataFrame into a NumPy array. 
scaler = StandardScaler()
market_data_scaled = scaler.fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
# Since market_data_scaled is a NumPy array, it has no column names. The following step creates a pandas DataFrame from the NumPy array.
# I reassigned the original columns so that the scaled data keeps the same column names. 
# I also reassigned the original index to the new DataFRame for consistency. 
scaled_market_data_df = pd.DataFrame(market_data_scaled, columns=df_market_data.columns)

# Copy the crypto names from the original DataFrame
crypto_names = df_market_data.index

# Set the coin_id column as index
scaled_market_data_df.index = crypto_names

# Display the scaled DataFrame
scaled_market_data_df.head(10)
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
# This step essentially creates a list of numbers, 1 - 10, that represent all the possible number of clusters to be tested when running KMeans. 
k_values = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for k in k_values:
    kmeans_model = KMeans(n_clusters=k, random_state=1)
    kmeans_model.fit(scaled_market_data_df)
    inertia.append(kmeans_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
# The dictionary consists of key value pairs: "k" is the list of k_values and "inretia" is the list of inertia values. 
# This step prepares the data in a structured way so that I can easily create the DataFrame for plotting. 
elbow_data = {"k": k_values, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
# While plotting the elbow_data, I assigned the plot to a variable so that I can later call upon it for contrasting with the other plot.  
original_elbow = elbow_df.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k_values)
original_elbow
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Looking at the above elbow curve, the best `k` value is 4. We can see that the inertia drops sharply from k=1 to k=4. After k4, the inertia has a flatter slpoe. This would mean that the data is well grouped into 4 clusters. Any additinal clusters would not improve the model other than making it more complex. 

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
# Based on the finding of the preceding elbow plot, 4 is decided to be the best value for k.
# Consequently, n_clusters=4 tells the model to form 4 clusters. 
# I used the random_state=1 to make the results reproducible. Not setting it could generate slightly different clusterings every time it is run. 
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled DataFrame
# By using .fit(), the KMeans model will be trained using the scaled data. 
model.fit(scaled_market_data_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
# This step will tell the trained KMeans model to assign each cryptocurrency in the scaled data to one of the four clusters. 
crypto_clusters = model.predict(scaled_market_data_df)

# Print the resulting array of cluster values.
print(crypto_clusters)
```

```python
# Create a copy of the scaled DataFrame
crypto_market_predictions = scaled_market_data_df.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
crypto_market_predictions["crypto_cluster"] = crypto_clusters

# Display the copy of the scaled DataFrame
crypto_market_predictions.head(10)
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
original_scatter = crypto_market_predictions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="crypto_cluster",
    hover_cols=["coin_id"],
    title="Clustered Cryptocurrencies by Price Change (24h vs 7d)"
)
original_scatter
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
# This step will reduce the original dataset down to just three principal components. 
# The goal is to apply a dimensionality reduction to compress the information in the data while keeping the most important patterns. 
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
# This step does two things: i learns the patterns in the scaled data and projects the data onto the three new directions set above. 
market_data_pca = pca.fit_transform(scaled_market_data_df)

# View the scaled PCA data
market_data_pca[:10]
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
# This step will show how important each PC(principal component) is. 
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the three principal components is 89.5%. Specifically, PC1 captures about 37.20%, PC2 captures 34.70%, and PC3 captures 17.60%. THis setup would mean that the three components preserve almost all of the important information. 

```python
# In this step, the goal is to turn the PCA results into a clean, labeled DataFrame with PC1, PC2, and PC3, and index it by cryptocurrency name. 
# Create a new DataFrame with the PCA data.
crypto_pca_df = pd.DataFrame(
    market_data_pca,
    columns=["PC1", "PC2", "PC3"]
)

# Copy the crypto names from the original scaled DataFrame
names_of_crypto = scaled_market_data_df.index

# Set the coin_id column as index
crypto_pca_df.index=names_of_crypto

# Display the scaled PCA DataFrame
crypto_pca_df.head(10)
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
pca_k_values = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
pca_inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in pca_k_values:
    pca_kmeans_model = KMeans(n_clusters=i, random_state=1)
    pca_kmeans_model.fit(crypto_pca_df)
    pca_inertia.append(pca_kmeans_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data = {"k":pca_k_values, "inertia": pca_inertia}

# Create a DataFrame with the data to plot the Elbow curve
pca_elbow_df = pd.DataFrame(pca_elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
# I assigned the plot to a variable to later call upon it for comparisons. 
pca_elbow = pca_elbow_df.hvplot.line(
    x="k",
    y="inertia",
    title="Reduced Crypto Data Elbow Curve",
    xticks=pca_k_values,
    color="red"
)
pca_elbow
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** As was the case before, the inertia drops sharply from k1 - k4 and the curve flattens out. Therefore, the best `k` after the PCA is still 4. 


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** The `k value` stays the same in both the original and the PCA-reduced data. This suggests that the clusters are strong and stable yet after dimentionality reduction. 

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
# The best value of k, which is 4 as shown in the elbow plot, is used for the n_clusters. 
pca_kmeans = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
pca_kmeans.fit(crypto_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
pca_crypto_clusters = pca_kmeans.predict(crypto_pca_df)

# Print the resulting array of cluster values.
print(pca_crypto_clusters)
```

```python
# Create a copy of the scaled PCA DataFrame
crypto_pca_predictions = crypto_pca_df.copy() 

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
crypto_pca_predictions["pca_crypto_clusters"] = pca_crypto_clusters

# Display the copy of the scaled PCA DataFrame
crypto_pca_predictions.head(10)
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
# I assigned the plot to a variable for later comparison.  
pca_scatter = crypto_pca_predictions.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="pca_crypto_clusters",
    hover_cols=["coin_id"],
    title="Cryptocurrencies Clustered by PCA Components (PC1 vs PC2)"
)
pca_scatter
```
